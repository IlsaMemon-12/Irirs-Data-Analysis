```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("iris")
```

```python
print("shape:", df.shape)
```

```python
print("Columns:", df.columns)
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
plt.figure()
sns.scatterplot(data=df, x="petal_width", y="petal_width", hue="species")
plt.title("Petal Length vs Petal Width")
plt.show()
```

```python
plt.figure()
sns.histplot(df["sepal_length"], kde=True)
plt.title("Distribution of sepal Length")
plt.show()
```

```python
plt.figure()
sns.boxplot(data=df)
sns.boxplot("Box Plot of Iris Features")
plt.show()
```
